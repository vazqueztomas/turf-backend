```python
from pypdf import PdfReader

def extract_text_from_pdf(pdf_path, page_number):
    # Create a PDF reader object
    reader = PdfReader(pdf_path)
    
    # Get the specified page
    page = reader.pages[page_number]
    
    # Extract text from the page
    text = page.extract_text()
    
    return text

# Example usage
pdf_path = "/home/tomasvazquez/Develop/turf-backend/playground/example.pdf"
page_number = 1  # Page numbers start from 0
text = extract_text_from_pdf(pdf_path, page_number)
print(text)
```

```python
from openai import OpenAI
client = OpenAI()

# Create a prompt for the OpenAI API
prompt = f"""
Extract all the races and the horses that run in each race from the following text. For each horse, include all relevant information such as name, number, jockey, trainer, last 5 races and any other details provided:

{text}

List all the races and the horses with their information in a structured format.
"""
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {
            "role": "user",
            "content": prompt
        }
    ]
)

print(completion.choices[0].message)
```
